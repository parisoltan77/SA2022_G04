# Convert .pkl to .csv

```python
import pickle
import os

with open('../Data/God_Class_metrics_and_heuristics.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
data
```

```python
import pandas as pd
import os
df = pd.DataFrame(data)
df.to_csv(os.path.join('../Data',r'God_Class_metrics_and_heuristics.csv'))
```
